```python
from framework import FrameWork
from framework import BorexinoRecoilSpectrum

from scipy.interpolate import interp1d
import re

from cobaya.run import run

from getdist.mcsamples import MCSamplesFromCobaya
import getdist.plots as gdplt

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['text.usetex']=True
plt.style.use('classic')

frame    = FrameWork(m12_nuisance=False)
```

```python
#Load and prepare the Data
def LoadData():
    d = np.zeros((817,7))
    with open('./Data/PRL2022_SubtractedSpectrumFit_Data.txt', "r") as text_file:
        for j,line in enumerate (text_file):
            a = re.findall(r"[-+]?(?:\d*\.*\d+)", line.split(", ")[0])
            for i in range(len(a)):
                d[j,i] = a[i]

    cond   = [250,510]
    n_th   = [5,6,10]
    d_rbin = [[],[],[],[],[]]

    for i in range (0,cond[0],n_th[0]):
        d_rbin[0].append(d[i,2])
        d_rbin[4].append(d[i,1])
        d_rbin[1].append(np.sum(d[i:i+n_th[0],3]))
        d_rbin[2].append(np.sum(d[i:i+n_th[0],4]))
        d_rbin[3].append(np.sqrt(np.sum(d[i:i+n_th[0],4])))   

    for i in range (cond[0],cond[1],n_th[1]):
        d_rbin[0].append(d[i,2])
        d_rbin[4].append(d[i,1])
        d_rbin[1].append(np.sum(d[i:i+n_th[1],3]))
        d_rbin[2].append(np.sum(d[i:i+n_th[1],4]))
        d_rbin[3].append(np.sqrt(np.sum(d[i:i+n_th[1],4])))   

    for i in range (cond[1],d.shape[0],n_th[2]):
        d_rbin[0].append(d[i,2])
        d_rbin[4].append(d[i,1])
        d_rbin[1].append(np.sum(d[i:i+n_th[2],3]))
        d_rbin[2].append(np.sum(d[i:i+n_th[2],4]))
        d_rbin[3].append(np.sqrt(np.sum(d[i:i+n_th[2],4])))  

    return np.array(d_rbin)

d_rbin = LoadData()

cond   = (d_rbin[0]>300)&(d_rbin[0]<600)
data   =  d_rbin[:,cond]
ydata  = data[2]
edata  = data[3]
xdata  = data[4]
```

```python
def Model(theta):
    dr_dldt     = frame[[theta,0]]

    elrec_be70  = BorexinoRecoilSpectrum(dr_dldt[0]['Be7'][0],frame.t_e['Be7'][0],frame.year,frame.theta)
    elrec_be71  = BorexinoRecoilSpectrum(dr_dldt[0]['Be7'][1],frame.t_e['Be7'][1],frame.year,frame.theta)

    nh_be7      = np.arange(90,950,1) 
    cond        = (nh_be7>=xdata[0])&(nh_be7<=xdata[-1])
    elrec_be7to = elrec_be70[cond]+elrec_be71[cond]
    nh_be7      = nh_be7[cond]

    theory_be7  = interp1d(nh_be7,elrec_be7to)
    return theory_be7(xdata)
model = Model(33.4)
```

```python
#Read the Background

def TetoNh(t_e):
     return -8.065244 + 493.2560*t_e - 64.09629*t_e**2 + 4.146102*t_e**3

def Po210Background():
    po210     = np.array(pd.read_csv('./Background/Po210_pdfkeV__11.txt',header = None))
    cond      = (po210[:,0]>300.)&(po210[:,0]<600.)
    po210_cut = po210[cond,:]
    nh        = TetoNh(1e-3*(po210_cut[:,0]+po210_cut[:,1])/2)-15
    return interp1d(nh,po210_cut[:,2])

po210_fit = Po210Background()
po210     = po210_fit(xdata)
```

```python
def LogLikelihood(flux,noise):
    prediction = flux*model+noise*po210
    return -np.sum((prediction-ydata)**2/edata**2)
    
#print(LogLikelihood(30,3000,4500))    

info = {"likelihood": {"Chi2": LogLikelihood},
        
        "params": dict([#("theta", {"prior": {"min": 10, "max": 50},
                        #          "latex": r"\theta_{12}"}),
                        ("flux" , {"prior": {"min": 0, "max": 1e4},
                                  "latex": r"\phi_{Be7}"}),
                        #("flux" , {"prior": {"dist": "norm", "loc": 4000, "scale":100},
                        #          "latex": r"\phi_{Be7}"}),
                        
                        ("noise" , {"prior": {"min": 0, "max": 1e4},
                                  "latex": r"\phi_{po210}"})
                        
                        ]),
        
        "sampler": {"mcmc": {"Rminus1_stop": 0.001, "max_tries": 100000}}}

updated_info, sampler = run(info)
```

```python
# Export the results to GetDist



gd_sample = MCSamplesFromCobaya(updated_info, sampler.products()["sample"])
  
gdplot = gdplt.get_subplot_plotter()
gdplot.triangle_plot(gd_sample, ["flux","noise"], filled=True)


gdplot.export('IPM3.pdf')
```

```python
mean   = gd_sample.getMeans()[:2]
covmat = gd_sample.getCovMat().matrix[:2, :2]
print("Mean:")
print(mean)
print("Covariance matrix:")
print(covmat)
```

```python
plt.figure(figsize=(20,5))
plt.rcParams.update({'font.size': 18})
plt.ylim(5e1,7e3)
plt.yscale('log')
plt.xlim(130,300)


plt.errorbar(xdata,ydata,edata,fmt='o')
plt.plot(xdata,mean[0]*model ,'g--',lw=2,label='Be7')
plt.plot(xdata,mean[1]*po210,'k--',lw=2,label='Po210')
plt.plot(xdata,mean[1]*po210+mean[0]*model,'r-',lw=2,label='Fit')


plt.xlabel(r'$N_h$')
plt.ylabel(r'$\rm Event$')

plt.legend(fontsize=14)
```
